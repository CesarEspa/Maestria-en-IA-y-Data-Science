```python
nota = 4.0
asistencia = 0.7 # 85%

umbral_nota = 4.0
umbral_asistencia = 0.8

if (nota >= umbral_nota) and (asistencia >= umbral_asistencia):
    resultado = "Aprobado y con buena asistencia"
elif nota >= umbral_nota:
    resultado = "Aprobo con baja asistencia"
else:
    resultado = "Reprobado"

print(f"El estudiante tuvo una nota de {nota}, una asistencia de => {asistencia*100}% y ha {resultado}")

```

```python
notas = [4.5, 4.2, 4.4, 3.1, 1.8, 2.5, 4.9, 3.7, 4.0, 2.2, 1.5, 3.3, 4.8, 2.9, 3.6, 4.1, 1.2, 5.0, 3.8, 
         2.7, 4.6, 3.0, 1.9, 4.3, 2.1, 3.4, 4.7, 1.6, 3.9, 2.8, 1.3, 3.2, 2.6, 4.0, 1.7, 3.5, 
         2.0, 4.2, 2.4, 1.1]

aprobados = []
reprobados = []
umbral = 4.0

for nota in notas:
    if nota >= umbral:
        aprobados.append(nota)
    else:
        reprobados.append(nota)

print(f"Total de notas {len(notas)}")
print(f"El numero de estudiantes aprobados es => {aprobados}")
print(f"El numero de estudiantes reprobados es => {reprobados}")
```
