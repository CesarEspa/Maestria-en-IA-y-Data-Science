```python
import pandas as pd
```

```python
df = pd.read_csv("Titanic-Dataset.csv")

df.head()
```

```python
# Obtener informaci√≥n del DataFrame, Aqui podemios ver el numero de filas, columnas y tipos de datos y datos que faltan en cada columna

df.info()
```

```python
# Contar los valores nulos en cada columna
df.isnull().sum()
```

```python
# Le sacamos la mediana a la edad para llenar los valores nulos

df['Age'].isnull().sum()

medianaAge = df["Age"].median()

print(medianaAge)

# Aqui remplezamos los datos nulos por la media de edad
df['Age'].fillna(medianaAge, inplace = True)
```

```python
# Ahora vemos que ya no hay datos nulos en AGE

df.isnull().sum()
```

```python
# Ahora vamos a seguir con los datos nulos con Cabin
# Entonces vamos a remplazar los valores nulos con desconocidos

df['Cabin'].fillna("Desconocido", inplace=True)
```

```python
df.isnull().sum()
```

```python
modaEmbarked = df['Embarked'].mode()

df["Embarked"].fillna(modaEmbarked[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
# Cuanta gente sobrevivio al titanic 

supervivientes = df[df["Survived"] == 1]

supervivientes.head()
```

```python
print(len(supervivientes))
```

```python
Sumasupervivientes = (df["Survived"] == 1).sum()

print(Sumasupervivientes)
```

```python
mujeres_sobrevivientes = df[(df["Survived"] == 1) & (df["Sex"] == "female")]

mujeres_sobrevivientes.head()
```

```python
print(len(mujeres_sobrevivientes))
```

```python

hombres_sobrevivientes = df[(df["Survived"] == 1) & (df["Sex"] == "male")]

hombres_sobrevivientes.head()
```

```python
jovenes_sobrevivientes = df[(df["Age"] >= 0) & (df["Age"] <= 17)]

print(len(jovenes_sobrevivientes))
```
